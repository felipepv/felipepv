```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Número de veículos pertencentes a cada marca


```python

grafico = sns.catplot(data=df,y="brand", kind="count", palette="rocket", height=8)
plt.xlabel("Número de veículos")
plt.ylabel("Marca")
grafico.ax.set_title("Veículos por Marca")
plt.show()
```

```python
# Salvando o plot
grafico.savefig(("plots/Analise2/brand-vehicleCount.png"))
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
fig = sns.catplot(data=df, kind="bar", x="vehicleType", y='price', hue="gearbox", palette="rocket", alpha=.6, height=10)
```

```python
# Salvando o plot
fig.savefig("plots/Analise2/vehicletype-gearbox-price.png")
```
