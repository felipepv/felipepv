```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
# Calcule a média de preço por marca e por veículo
```

```python
avgPrice = pd.DataFrame()
for a in list(df['brand'].unique()):
    for b in list(df['vehicleType'].unique()):
        c = df[(df['brand'] == a) & (df['vehicleType']== b)]["price"].mean()
        avgPrice = avgPrice.append(pd.DataFrame({'brand':a, 'vehicleType':b, 'avgPrice':c}, index=[0]))
```

```python
avgPrice.head()
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Crie um Heatmap com Preço médio de um veículo por marca, bem como tipo de veículo
```

```python
tres = avgPrice.pivot("brand","vehicleType","avgPrice")
fig, ax = plt.subplots(figsize=(15,20))
sns.heatmap(tres,linewidths=1, cmap="jet",annot=True, ax=ax, fmt="d")
plt.show()
```
